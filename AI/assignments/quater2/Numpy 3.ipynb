#### Derivatives can be calculated numerically with the finite difference method as:
f'(x)=f(x+del(x)) - f(x-del(x))/2*del(x)

Construct 1D Numpy array containing the values of x in the interval [0,π/2] with spacing Δx=0.1. Evaluate numerically
the derivative of sin in this interval (excluding the end points) using the above formula. Try to avoid for loops.
Compare the result to function cos in the same interval

```python
import numpy as np
import matplotlib.pyplot as plt

# Construct the 1D array of x values
dx = 0.1
x = np.arange(0, np.pi/2, dx)
print(x)

# Calculate the numerical derivative of sin(x) using the finite difference method
sin_x = np.sin(x)
print(np.roll(sin_x, -1))
sin_dx = (np.sin(x + 0.1) - np.sin(x - 0.1)) / (2 * 0.1)# (np.roll(sin_x, -1) - np.roll(sin_x, 1))/(2*dx)
print(sin_dx)
sin_dx = sin_dx[1:-1] # exclude the end points

# Calculate the actual derivative of cos(x)
cos_dx = -np.cos(x)[1:-1]

# Plot the results
plt.plot(x[1:-1], sin_dx, label='Numerical derivative of sin(x)')
plt.plot(x[1:-1], cos_dx, label='Derivative of cos(x)')
plt.legend()
plt.show()
```
